# 📘 بخش سوم: تابع هزینه و بهینه‌سازی در یادگیری ماشین

## 🧠 چرا تابع هزینه؟تابع هزینه به ما می‌گوید مدل چقدر خوب یا بد پیش‌بینی می‌کند.هدف ما کمینه‌سازی این تابع است.

## 🎯 تابع هزینه MSE\[J(w) = \frac{1}{n} \sum_{i=1}^{n} \left( h(x_i) - y_i \right)^2\]

## 🔍 تابع هزینه MAE\[J(w) = \frac{1}{n} \sum_{i=1}^{n} |h(x_i) - y_i|\]

## 📉 بهینه‌سازی (Optimization)پیدا کردن پارامترهایی مثل w که تابع هزینه را کمینه کنند.

## ⚙️ گرادیان نزولی\[w := w - \eta \cdot \nabla J(w)\]

```python
# 📌 شبیه‌سازی گرادیان نزولی روی تابع ساده J(w) = (w - 3)^2import numpy as npimport matplotlib.pyplot as pltdef J(w): return (w - 3)**2           # تعریف تابع هزینهdef grad_J(w): return 2 * (w - 3)     # مشتق تابع هزینهeta = 0.1        # نرخ یادگیریw = 0            # مقدار اولیهhistory = [w]    # ذخیره مسیر حرکتfor i in range(20):    w = w - eta * grad_J(w)    history.append(w)w_vals = np.linspace(-1, 6, 100)j_vals = J(w_vals)plt.plot(w_vals, j_vals, label='J(w)')plt.scatter(history, [J(w) for w in history], color='red', label='مراحل GD')plt.title('گرادیان نزولی روی تابع J(w) = (w - 3)^2')plt.xlabel('w')plt.ylabel('J(w)')plt.legend()plt.grid(True)plt.show()
```

## ⚠️ نرخ یادگیری مناسب؟| مقدار η (eta) | رفتار ||---------------|--------|| خیلی کوچک     | کند و آهسته || خیلی بزرگ     | نوسانی و ناپایدار || مناسب         | سریع و همگرا |
